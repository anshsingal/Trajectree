```python
import networkx as nx
from networkx.drawing.nx_pydot import graphviz_layout
graph = nx.DiGraph()
```

```python
graph.add_node((0,0,1))
graph.add_edge((0,0,1), (0,0,1,2))
```

```python
# You can either use the default layouts in NetworkX:  
pos = nx.spring_layout(graph) 

# For a pretty tree like graph, you need to have Graphviz installed:
# pos = graphviz_layout(graph, prog="dot") # You need Graphviz installed for this to work
```

```python
nx.draw(graph, pos, with_labels=False, node_color='pink', node_size=500, alpha=0.9)
node_descriptions = None # {(0,0,1): "Root Node", (0,0,1,2): "Child Node"}
print(node_descriptions)
nx.draw_networkx_labels(graph, pos, labels=node_descriptions, font_size=12, font_color='black')
```
